```python
import json
from datetime import datetime

MODEL="gpt-3.5-turbo"

prompts = [
    "I can't focus on anything today, what should I do?",
    "How do I manage my time better with ADHD?",
    "I keep forgetting important tasks, any tips?"
]

def ask_prompts(model_name, label):
    results = []
    print(f"\n=== {label} (model: {model_name}) ===\n")
    for prompt in prompts:
        response = client.chat.completions.create(
            model=model_name,
            messages=[
                {"role": "system", "content": "You are an ADHD support assistant."},
                {"role": "user", "content": prompt}
            ]
        )
        answer = response.choices[0].message.content
        print(f"Q: {prompt}")
        print(f"A: {answer}")
        print("-" * 50)
        results.append({
            "prompt": prompt,
            "response": answer
        })
    return results

def save_comparison(before, after, path="./data/comparison.json"):
    output = {
        "timestamp": datetime.now().isoformat(),
        "comparisons": [
            {
                "prompt": b["prompt"],
                "before": b["response"],
                "after": a["response"]
            }
            for b, a in zip(before, after)
        ]
    }
    with open(path, "w", encoding="utf-8") as f:
        json.dump(output, f, indent=2, ensure_ascii=False)
    print(f"\nComparison saved to {path}")

# --- STEP 1: Run before fine-tuning ---
before_results = ask_prompts(MODEL, "BEFORE FINE-TUNING")
```

```python
import matplotlib.pyplot as plt

def get_all_metrics(job_id):
    """Pull all metrics events from fine-tuning job"""
    all_events = []
    after = None

    # paginate through all events
    while True:
        events = client.fine_tuning.jobs.list_events(
            job_id,
            limit=100,
            after=after
        )
        all_events.extend(events.data)
        if not events.has_more:
            break
        after = events.data[-1].id

    # extract metrics
    steps = []
    training_loss = []
    validation_loss = []
    training_accuracy = []
    validation_accuracy = []

    for event in reversed(all_events):
        if event.type == "metrics":
            m = event.data
            if m.get("step") is not None:
                steps.append(m["step"])
            if m.get("train_loss") is not None:
                training_loss.append(m["train_loss"])
            if m.get("valid_loss") is not None:
                validation_loss.append(m["valid_loss"])
            if m.get("train_mean_token_accuracy") is not None:
                training_accuracy.append(m["train_mean_token_accuracy"])
            if m.get("valid_mean_token_accuracy") is not None:
                validation_accuracy.append(m["valid_mean_token_accuracy"])

    return {
        "steps": steps,
        "training_loss": training_loss,
        "validation_loss": validation_loss,
        "training_accuracy": training_accuracy,
        "validation_accuracy": validation_accuracy,
    }


def plot_dashboard_metrics(metrics):
    """Recreate OpenAI dashboard graphs in notebook"""
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    fig.suptitle("Fine-Tuning Metrics", fontsize=14, fontweight="bold")

    steps = metrics["steps"]

    # --- Training and Validation Loss ---
    ax1 = axes[0]
    if metrics["training_loss"]:
        ax1.plot(steps[:len(metrics["training_loss"])],
                 metrics["training_loss"],
                 label="Training Loss", color="blue", marker="o", markersize=3)
    if metrics["validation_loss"]:
        ax1.plot(steps[:len(metrics["validation_loss"])],
                 metrics["validation_loss"],
                 label="Validation Loss", color="orange", marker="o", markersize=3)
    ax1.set_xlabel("Step")
    ax1.set_ylabel("Loss")
    ax1.set_title("Training vs Validation Loss")
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    # --- Training and Validation Token Accuracy ---
    ax2 = axes[1]
    if metrics["training_accuracy"]:
        ax2.plot(steps[:len(metrics["training_accuracy"])],
                 metrics["training_accuracy"],
                 label="Training Accuracy", color="green", marker="o", markersize=3)
    if metrics["validation_accuracy"]:
        ax2.plot(steps[:len(metrics["validation_accuracy"])],
                 metrics["validation_accuracy"],
                 label="Validation Accuracy", color="red", marker="o", markersize=3)
    ax2.set_xlabel("Step")
    ax2.set_ylabel("Token Accuracy")
    ax2.set_title("Training vs Validation Token Accuracy")
    ax2.legend()
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    # plt.savefig("./data/dashboard_metrics.png", dpi=150)
    plt.show()
    # print("Saved to ./data/dashboard_metrics.png")
```

```python
def print_summary(metrics):
    """Print final metric values"""
    print("\n=== Metrics Summary ===")
    if metrics["training_loss"]:
        print(f"Final training loss:       {metrics['training_loss'][-1]:.4f}")
    if metrics["validation_loss"]:
        print(f"Final validation loss:     {metrics['validation_loss'][-1]:.4f}")
    if metrics["training_accuracy"]:
        print(f"Final training accuracy:   {metrics['training_accuracy'][-1]:.4f}")
    if metrics["validation_accuracy"]:
        print(f"Final validation accuracy: {metrics['validation_accuracy'][-1]:.4f}")
```

                          Prompt Experiment Types vs Prompting in Generative AI

Prompt Experimentation Prompt experimentation is the systematic process of testing and varying prompt components to measure their effect on model output quality. It treats prompting as an empirical discipline — changing one variable at a time, such as phrasing, examples, temperature, or instruction specificity, and observing how outputs shift. Experiments might test whether adding "think step by step" improves reasoning accuracy, whether a system prompt outperforms a user-turn instruction, or whether few-shot examples reduce hallucination rates. The goal is not just a good output, but knowledge about which prompt design reliably produces good outputs across varied inputs. In production AI systems, this methodology separates robust pipelines from brittle ones.

Prompting Prompting is the practice of crafting input text to guide a generative AI model toward a desired output. Techniques include zero-shot prompting, which provides no examples; few-shot prompting, which embeds input-output examples directly in the prompt; chain-of-thought prompting, which instructs the model to reason step by step; and role prompting, which assigns a persona to shape tone and expertise. Prompting is the primary interface between humans and LLMs, and its effectiveness depends on structure, context, and intent.
