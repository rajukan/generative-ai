```python
def _set_env_from_file(var: str, file_path: str = r"C:\Users\gyanr\gyan-python-workspace\grk_langchain\app\notebooks\data\openai_key.txt"):
    """
    Reads an API key from a specified file and sets it as an environment variable.
    """
    if not os.environ.get(var):
        try:
            # The 'with open' statement ensures the file is closed automatically
            with open(file_path, 'r') as f:
                # Read the first line and strip any leading/trailing whitespace
                key = f.readline().strip()

            if key:
                os.environ[var] = key
                print(f"Successfully loaded {var} from {file_path}")
            else:
                print(f"Warning: {file_path} is empty.")

        except FileNotFoundError:
            print(f"Error: Key file not found at {file_path}. Please create the file.")

```

#### Checks and Balances

Ensure the basic checks and formatting of the file are inplace even before callign the api

```python
def basic_checks(filename):
    with open(filename, "r") as f:
        lines = f.readlines()
    print(f"Found {len(lines)} examples in {filename}")
    print("First example preview:")
    print(lines[0])

def format_checks(filename):
    with open(filename, "r") as f:
        for i, line in enumerate(f, 1):
            try:
                obj = json.loads(line)
                assert "messages" in obj, "Missing 'messages' key"
            except Exception as e:
                print(f"Format error on line {i}: {e}")
                return
    print("Format checks passed")

# Run checks
basic_checks(training_data_file)
format_checks(training_data_file)
```

# Upload

Once the file check is passed, next would be to call the openai client and upload it for fine-tuning.

## Fine Tune Model

Start fine-tuning using gpt-3.5-turbo as the base model

```python
job = client.fine_tuning.jobs.create(
    training_file=file.id,
    model="gpt-3.5-turbo",
    suffix="emoji-bot"
)

print("Fine-tuning job started:")
print(job)
print(f"Job ID: {job.id}")
```

## Fire and Forget Alternative approach

If you want to "fire and forget" your OpenAI fine-tuning job—meaning you submit the job and let it run in the background without actively polling for status—you can simplify your code to just submit the job and optionally log the job ID for later reference. Here’s how:
make sure you capture the job id from previous trigger

```python
JOB_ID = job.id
```

```python
# Retrieve the job status
ft_job = client.fine_tuning.jobs.retrieve(JOB_ID)
status = ft_job.status

print(f"Job Status: {status}")
if status == 'succeeded':
    print(f"Fine-tuned model: {ft_job.fine_tuned_model}")
elif status == 'failed':
    print(f"Error: {ft_job.error}")
else:
    print(f"Job is still {status}. Check back later.")
```

Key Points

Fire-and-Forget: The job runs in the background on OpenAI’s servers. You don’t need to keep your script or notebook running.  
Job ID: Always log the job.id so you can check the status later.  
No Polling: You avoid the while loop and time.sleep calls, which means your script exits immediately after submitting the job.  
Manual Check: Use the second code block whenever you want to check the job status.  


Incase job fails, below can be helpful

```python
if status == 'failed':
    print("Fine-tuning job failed.")
    print(f"Error code: {ft_job.error.code}")
    print(f"Error message: {ft_job.error.message}")
    print(f"Full job details: {ft_job}")
```
